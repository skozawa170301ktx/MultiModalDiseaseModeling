# 3_path_id.ipynb
Convert the ID in the 'order' column, which is specific to each hsa_map ID, to a KEGG ID. (except hsa00140, hsa00240, hsa00250, hsa00564, hsa00590 and hsa00760)
### input
- 1_Paths_from_KEGG_Pathway_code/output/2_paths_extraction.csv : A file linking Target and Path which contain IDs in the 'order' column, which is specific to each hsa_map ID
- 1_Paths_from_KEGG_Pathway_code/output/all_target_node_DrugBank.csv : A file with nodes information from KEGG
### output
- 2_Paths_from_KEGG_Pathway_datafile/Paths_from_KEGG_Pathway.csv : A file with Path, Pathway map ID and Target associated
- 1_Paths_from_KEGG_Pathway_code/output/all_target_node_DrugBank_group.csv : Added a summary of the group's KEGG IDs to 'all_target_node_DrugBank.csv'

```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv('output/2_paths_extraction.csv',header = 0, index_col=0)
df['order'] = df['order'].apply(eval)
```

```python
df_node = pd.read_csv('output/all_target_node_DrugBank.csv',header = 0, index_col=0)
df_node = df_node.rename(columns={'index':'ID'})
```

```python
df_a = df.reset_index()
df_a = df_a.drop(columns = ['level_0','level_1'])
```

```python
df_a
```

```python
df_a_unstack = df_a['order'].apply(pd.Series).unstack().reset_index().sort_values(['level_1','level_0']).dropna(subset=[0]).reset_index(drop = True)
df_c = pd.merge(df_a_unstack,df_a,left_on='level_1',right_on ='index')
df_c = df_c[['index', 'HSA', 'hsa_map', 'order','level_0', 0]]
df_c  = df_c.rename(columns = {'level_0':'index_m'})
df_c  = df_c.rename(columns = {'order':'order1'})
df_c  = df_c.rename(columns = {0:'order'})
```

```python
df_c
```

df_n2 : Added a summary of the group's KEGG IDs to df_node.

```python
def list_to_str(list_df):
    return ' '.join(list_df)
```

```python
def str_to_list(list_df):
    return list_df.split(' ')
```

```python
df_n2 = pd.DataFrame()
for i in list(set(df_node['hsa'])):
    d = df_node[df_node['hsa']==i]
    df_gene = d[d['type'] != 'group']
    df_group = d[d['type'] == 'group']
    if len(df_group) == 0:
        df_n2 = pd.concat([df_n2, df_gene[['ID','hsa','KEGG_id']]])
    else:
        df_group['component'] = df_group['component'].apply(eval)
        c = df_group['component'].apply(pd.Series).unstack().reset_index().sort_values(['level_1','level_0']).dropna(subset=[0]).reset_index(drop = True)
        df_g = pd.merge(c,df_group,left_on='level_1',right_index =True)
        df_g[0] = df_g[0].astype(int)
        df_g1 = pd.merge(df_g[[0,'ID','hsa']], df_gene[['ID','KEGG_id','hsa']], left_on = [0,'hsa'], right_on = ['ID','hsa'])
        df_g1['KEGG_id'] = df_g1['KEGG_id'].apply(str_to_list)
        df_g2 = df_g1.groupby(['ID_x', 'hsa'])['KEGG_id'].apply(lambda x: [s for s in x.values.tolist()]).reset_index()
        df_g2['KEGG_id'] = df_g2['KEGG_id'].apply(itertools.chain.from_iterable).apply(list)
        df_g2['KEGG_id'] = df_g2['KEGG_id'].apply(sorted)
        df_g2 = df_g2.rename(columns = {'ID_x':'ID'})
        df_g2['KEGG_id'] = df_g2['KEGG_id'].apply(list_to_str)
        df_g3 = pd.concat([df_gene[['ID','hsa','KEGG_id']],df_g2])
        df_n2 = pd.concat([df_n2, df_g3])
df_n2 = df_n2.reset_index(drop = True)
```

```python
df_n2
```

```python
df_n2.to_csv('output/all_target_node_DrugBank_group.csv',encoding = 'utf-8')
```

For each hsa_map number, merge df_n2 into df_c.

```python
df_all = pd.DataFrame()
for i in list(set(df_c['hsa_map'])):
    d = df_c[df_c['hsa_map']==i]
    e = df_n2[df_n2['hsa']==i]
    f = pd.merge(d, e, left_on = 'order',right_on = 'ID', how = 'left')
    df_all = pd.concat([df_all, f])
df_all = df_all[['index', 'HSA', 'hsa_map', 'order1', 'ID', 'index_m','KEGG_id']]
df_all = df_all.sort_values(['index', 'index_m'])
df_all.index = list(df_all['index_m'])
```

```python
df_all
```

```python
df_all['order1'] = df_all['order1'].astype(str)
```

```python
df_name = df_all.groupby(['index', 'HSA', 'hsa_map', 'order1'])['KEGG_id'].apply(lambda x: [s for s in x.values.tolist()]).reset_index()
```

```python
df_name['KEGG_id'] = df_name['KEGG_id'].astype(str)
```

Excluding hsa00140, hsa00240, hsa00250, hsa00564, hsa00590 and hsa00760

```python
df_name = df_name[df_name['hsa_map'] != 'hsa00140']
df_name = df_name[df_name['hsa_map'] != 'hsa00240']
df_name = df_name[df_name['hsa_map'] != 'hsa00250']
df_name = df_name[df_name['hsa_map'] != 'hsa00564']
df_name = df_name[df_name['hsa_map'] != 'hsa00590']
df_name = df_name[df_name['hsa_map'] != 'hsa00760']
```

```python
df_name = df_name.drop(columns = ['index']).sort_values('KEGG_id').reset_index(drop = True)
```
