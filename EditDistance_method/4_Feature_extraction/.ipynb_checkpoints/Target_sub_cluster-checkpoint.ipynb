```python
df_t = pd.read_csv('../2_Paths_from_KEGG_Pathway_datafile/Paths_from_KEGG_Pathway.csv',header = 0, index_col=0)
```

```python
df_t = df_t[['HSA', 'KEGG_id']].drop_duplicates().sort_values('HSA').groupby(['KEGG_id'])['HSA'].apply(list).reset_index().sort_values('KEGG_id')
```

```python
df_t['HSA'] = df_t['HSA'].apply(str)
```

```python
df_hsa = pd.merge(df_order, df_t, on = 'KEGG_id')
```

```python
df_hsa
```

```python
df_hsa[['index', 'HSA']].groupby(['HSA']).count().sort_values('index')
```

```python
df = pd.read_csv('../3_Calc_Edit_Distance/output/Calc_Edit_Distance.csv.gz', header = 0, index_col=0)
df_same = pd.DataFrame(range(67481), columns = ['down_index_new_0'])
df_same['down_index_new_1'] = df_same['down_index_new_0']
df_same['value'] = 1
df = pd.concat([df, df.rename(columns = {'down_index_new_0':'a'}).rename(columns = {'down_index_new_1':'down_index_new_0'}).rename(columns = {'a':'down_index_new_1'})])
df = pd.concat([df, df_same])
```

```python
df.head()
```

```python
df_cluster = pd.DataFrame()
for target_id in list(set(df_hsa['HSA'])):
    df_map = df_hsa[df_hsa['HSA'] == target_id][['index']]
    if len(df_map) != 1:
        df_y = pd.merge(df, df_map, left_on = 'down_index_new_0', right_on = 'index').drop(columns = 'index')
        df_y['id_0'] = df_y['down_index_new_0'].rank(method='dense') - 1
        df_y['id_0'] = df_y['id_0'].astype(int)
        
        df_c = pd.DataFrame(csr_matrix((list(df_y['value']), (list(df_y['down_index_new_1']), list(df_y['id_0']))), shape=(67481, df_y['id_0'].max() + 1)).toarray())
        distance_matrix  = 1 - np.abs(df_c.corr())
        dist_linkage  = hierarchy.average(squareform(distance_matrix))
        cluster_ids = hierarchy.fcluster(dist_linkage, 0.2, criterion="distance")
        df1 = pd.DataFrame(cluster_ids, distance_matrix.index).rename(columns = {0:'sub_ID'}).reset_index()
        df1 = pd.merge(df1, df_y[['down_index_new_0', 'id_0']].drop_duplicates().reset_index(drop = True), left_on = 'index', right_on = 'id_0').drop(columns = 'id_0')
        df1['HSA'] = target_id
        df_cluster = pd.concat([df_cluster, df1])
    else:
        df1 = df_map[['index']].rename(columns = {'index':'down_index_new_0'})
        df1['index'] = 0
        df1['sub_ID'] = 1
        df1['HSA'] = target_id
        df_cluster = pd.concat([df_cluster, df1])
df_cluster = df_cluster.reset_index(drop = True)
```
